```python

# Data for the pie chart
categories = [
    "Withdrawn or \n Desk Rejected",
    "Rejected",
    "Accepted as Poster",
    "Accepted as Oral",
]
values = [991, 3719, 1674, 154]
total = 6538
percentages = [round(v / total * 100, 2) for v in values]

# Create labels with percentages
plt.figure()

tran = 0.85
colors = [(1, 0, 0, tran),  # Red with 90% opacity
          (30/255, 144/255, 255/255, tran),  # DodgerBlue with 90% opacity
          (144/255, 238/255, 144/255, tran),  # LightGreen with 90% opacity
          (1, 215/255, 0, tran)]  # Gold with 90% opacity

plt.pie(values, autopct='%1.1f%%', startangle=210, colors=colors, pctdistance=0.7)
plt.legend(categories, loc="upper right")
plt.savefig(folder_path + "pie_chart_acceptance.pdf", bbox_inches='tight')
plt.show()
```

```python

# Data for the pie chart
categories = ["Very unlikely", "Somewhat unlikely", "Maybe", "Somewhat likely", "Very likely"]
values = [0.034, 0.022, 0.132, 0.218, 0.594]

# Create labels with percentages
plt.figure()
tran = 1
colors = [(1, 0, 0, tran),  # Red (80% opacity)
          (30/255, 144/255, 255/255, tran),  # DodgerBlue (80% opacity)
          (0, 1, 0.6, tran),  # SpringGreen (80% opacity)
          (1, 215/255, 0, tran),  # Gold (80% opacity)
          (1, 0.65, 0, tran)]  # Orange (80% opacity)
wedges, texts, autotexts = plt.pie(values, colors=colors,
                                   startangle=100, autopct='%1.1f%%', pctdistance=0.7)

# Adjust only the 2.0% label
for i, autotext in enumerate(autotexts):
    if values[i] == 0.022:  # 2.0% value
        autotext.set_position((autotext.get_position()[0] * 1.3, autotext.get_position()[1] * 1.3))

plt.legend(wedges, categories, loc="lower right")
plt.savefig(folder_path + "pie_chart_likelihood.pdf", bbox_inches='tight')
plt.show()
```
